```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = "/content/Data_Tanaman_Padi_Sumatera_version_1.csv"
df = pd.read_csv(dataset)
df.head()
```

```python
sns.countplot( x = df['Kelembapan'] )
```

```python
sns.distplot(df['Curah hujan'])
```

```python
sns.boxplot(df['Luas Panen'])
```

```python
sns.scatterplot(x= df['Produksi'], y = df['Luas Panen']  , hue = df['Curah hujan'] , style = df['Kelembapan'] , size = df['Suhu rata-rata'])
plt.show ()
df['Produksi'].min()
```

```python
sns.barplot (x = df['Produksi'] , y = df['Luas Panen'] , hue = df['Curah hujan'])
```

```python
sns.boxplot( x = df['Produksi'] , y = df['Kelembapan'] , hue = df['Suhu rata-rata'] )
```

```python
import warnings
warnings.filterwarnings('ignore')
df['Produksi'].mean()

sns.distplot( df[df['Luas Panen'] ==  'Right' ]['Produksi'], hist = False)
sns.distplot( df[df['Luas Panen'] ==  'Left' ]['Produksi'] , hist = False)
```

```python
df['Produksi'] = pd.to_numeric(df['Produksi'], errors='coerce')
print(df['Produksi'].dtype)

skill_X_preferredfoot  = pd.crosstab( df['Produksi'], df['Luas Panen'])
sns.heatmap(skill_X_preferredfoot)
plt.show()

new = df.groupby('Luas Panen')['Produksi'].mean()
new.plot(kind='bar')
plt.show()
```

```python
cluster = pd.crosstab(df['Produksi'] ,df['Kelembapan'] )
sns.clustermap(cluster)
```
